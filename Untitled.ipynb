```python
import pandas  as pd
```

```python
df=pd.read_csv("./Heart_Disease_Prediction.csv")
```

```python
df
```

```python
df.columns
```

```python
df.columns = (
    df.columns
    .str.lower()
    .str.replace(" ", "_")
)
```

```python
df.isnull().sum()
```

```python
df.dtypes
```

```python
columns=df.columns
for i in columns:
    print(i)
    print(df[i].unique())
```

```python
df['heart_disease'] = df['heart_disease'].map({
    'Presence': 0,
    'Absence': 1
})
```

```python
df['heart_disease'] 
```

```python
df['heart_disease'] = df['heart_disease'].astype(int)
```

```python
df.dtypes
```

```python
df.to_csv("heart_data_clean.csv", index=False)
```
